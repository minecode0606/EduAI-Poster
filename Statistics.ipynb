```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
school_data = pd.read_csv("/content/School_utf8.csv")
```

```python
region = list(school_data["소속지역"])
```

```python
# 각 숫자의 빈도수를 저장하는 딕셔너리를 초기화
frequency = {}
for i in region:
    try:
        frequency[i] += 1
    except:
        frequency[i] = 0
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터를 값에 따라 내림차순으로 정렬
sorted_data = dict(sorted(frequency.items(), key=lambda item: item[1], reverse=True))

# 데이터의 키와 값을 분리
labels = list(sorted_data.keys())
sizes = list(sorted_data.values())
total = sum(sizes)

# 바 차트를 그립니다
fig, ax = plt.subplots(figsize=(10, 6))  # 사이즈는 적절히 조절
sns.barplot(x=labels, y=sizes, ax=ax, palette='viridis')

# 각 막대의 위에 개수와 비율을 표시합니다
for i, value in enumerate(sizes):
    ax.text(i, value, f"{value} \n ({value / total:.1%})", ha='center', va='bottom')

plt.title("AI교육 학교의 지역별 비중", fontsize=27)
plt.xticks(rotation=90)  # 라벨이 겹치는 것을 방지하기 위해 x축 라벨을 회전시킵니다
plt.tight_layout()

plt.show()
```

```python
 surveydata = pd.read_csv("/content/AI_survey.csv")
 surveydata.head()
```

```python
good_AI_study = list(surveydata["귀하의 학교나 교육기관에서 인공지능 교육이 잘 이뤄지고 있다고 생각하나요?"])
satisfy_AI = list(surveydata["AI 교육의 필요성을 얼마나 느끼시나요?\n(AI개발, AI사용 시 윤리적 책임 등 전반적인 교육)"])
```

```python
ye = 0
no = 0
for i in zip(good_AI_study, satisfy_AI):
    if (i[0] == "매우 그렇다" or i[0] == "그렇다") and (i[1] == "인공지능 교육은 필수적으로 진행되어야 한다" or "인공지능 교육은 선택적으로 진행되어야 한다"):
        ye += 1
    else:
        no += 1
```

```python
print(ye)
print(no)
print(ye + no)
```

```python
# 두 숫자를 리스트로 만듭니다
numbers = [ye,no]

# 원형 차트를 그립니다
plt.pie(numbers, labels=["인공지능교육에 만족한다", "인공지능교육에 만족하지 않는다"], autopct='%1.1f%%', textprops={'fontsize': 17})
plt.title("인공지능 교육에 대한 유경험 학생들의 통계", fontsize=25, y=1.05)

# 차트를 화면에 보여줍니다
plt.show()
```

```python
ai_use_data = list(surveydata["당신은 chat gpt, PaLM2(Bard), Bing chat 등 생산적 대화형 모델을 얼마나 자주 사용하나요?"])
use = 0
no_use = 0
for i in ai_use_data:
    if i == "사용하지 않음":
        no_use += 1
    else:
        use += 1

# 두 숫자를 리스트로 만듭니다
numbers = [use, no_use]

# 원형 차트를 그립니다
plt.pie(numbers, labels=["사용한다", "사용하지 않는다"], autopct='%1.1f%%', textprops={'fontsize': 17})
plt.title("대화형 AI 사용 빈도수", fontsize=25, y=1.05)

# 차트를 화면에 보여줍니다
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# 데이터 정의
data = {'Label': ['AI교육을 통해 \n컴퓨터를 더욱 잘 이해할\n 수 있었다', 'AI교육을 통해 \n컴퓨터에 관심이\n 생겼다', 'AI교육을 통해 \nAI를 더욱 잘 이해하고\n 사용할 수 있었다.','AI교육을 통해 \n창의적인\n 문제해결능력을 \n키울 수 있었다.','AI교육을 통해\n 다양한 분야에 응용할\n 수 있었다.'], 'Value': [11,13,11,2,8]}
total = sum(data['Value'])  # 비율 계산을 위한 총합 계산

# DataFrame 생성
df = pd.DataFrame(data)

# 비율 컬럼 추가
df['Percentage'] = df['Value']/total*100

# 그래프 그리기
plt.figure(figsize=(8, 4))
barplot = sns.barplot(data=df, x='Label', y='Value', palette='viridis')

plt.title("AI교육을 통해 느낀 좋은 점")
# 막대에 개수 및 비율 추가
for i, row in df.iterrows():
    barplot.text(i, row.Value/2, f"갯수: {row.Value},\n비율: {row.Percentage:.2f}%", color='white', ha='center', va='center')

# 그래프 보이기
plt.show()
```
